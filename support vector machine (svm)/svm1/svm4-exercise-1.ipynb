```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits

digits = load_digits()
```

```python
dir(digits)
```

```python
digits.feature_names
```

```python
df = pd.DataFrame(digits.data, columns=digits.feature_names)
df.head()
```

```python
df['target'] = digits.target
df.head()
```

```python
df['digit_name'] = df.target.apply(lambda x: digits.target_names[x])
df.head()
```

```python
df0 = df[df.target == 0]
df1 = df[df.target == 1]
df2 = df[df.target == 2]
df3 = df[df.target == 3]
df4 = df[df.target == 4]
```

```python
plt.title('pixel_0_0 vs pixel_0_1 for df0, df1, df2, df3 and df3 data frame')
plt.xlabel('pixel_0_0')
plt.ylabel('pixel_0_1')
plt.scatter(df0['pixel_0_0'], df0['pixel_0_1'], color='black')
plt.scatter(df1['pixel_0_0'], df1['pixel_0_1'], color='green')
plt.scatter(df2['pixel_0_0'], df2['pixel_0_1'], color='blue')
plt.scatter(df3['pixel_0_0'], df3['pixel_0_1'], color='orange')
plt.scatter(df4['pixel_0_0'], df4['pixel_0_1'], color='red')
```
