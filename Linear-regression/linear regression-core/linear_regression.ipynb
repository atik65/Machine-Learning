```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('train_data.csv')
df
```

```python
x = df[['area']]
y = df['population']
```

```python
plt.scatter(x,y)
plt.title('Linear Regression: Area vs. Population')
plt.xlabel('Area in sq KM')
plt.ylabel('Population')
```

```python
model = LinearRegression()
model.fit(x,y)
y_predict = model.predict(x)
```

```python
plt.scatter(x,y)
plt.title('Linear Regression: Area vs. Population')
plt.xlabel('Area in sq KM')
plt.ylabel('Population')
plt.plot(x,y_predict)
```

```python
testing_data_frame = pd.read_csv('testing_data.csv')
y_predict_for_testing_data = model.predict(testing_data_frame)

# saving predicted data to new csv
testing_data_frame['population'] = y_predict_for_testing_data
testing_data_frame.to_csv('predicted_data.csv', index=False)

```
